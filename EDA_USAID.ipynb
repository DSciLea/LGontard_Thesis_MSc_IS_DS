# Data context

The aim of this project is to use Natural Language Processing (NLP) techniques, specifically using pre-trained Bidirectional Encoder Representations from Transformers (BERT) and Robustly Oprtimized BERT Pretraining Approach (RoBERTa)models, to measure the degree of politicization in USAID public aid project evaluation reports (across different political administrations).

With this research, we aim to contribute to the research gap that lies in the absence of quantitative scores able to model variable degrees of politicization for a topic given different contexts. By using contextualized embedding as a novelty, we address the limitation of the state-of-the-art that develops a framework to compute politicization scores for reports using static embedding. Pretrained contextual embedding provides a more granular analysis of politicization using contextualization.

**USAID’s Development Experience Clearinghouse (DEC) reports:**


Our main dataset focuses on health projects publicly available through the USAID's Development Experience Clearinghouse (DEC) (see https://dec.usaid.gov/dec/home/Default.aspx). We retrieve health-related reports following the hypothesis that politicization is expected to vary across sectors and activities, and more precisely that reproductive and maternal healthcare is more politicized than other sectors. Among all the reports, it contains 4,000 evaluations on health spanning over 70 years, from which we selected a balanced sample of 99 reports following a previous study on politicization (Bloem, 2024). The sample is considered balanced regarding the representation of years and countries related to the projects. They were written between 2000 and 2021 by third-party evaluators, following the guidelines of the USAID Evaluation Reports paper. All reports are written in English but one, and most of the reports were conducted on the African continent, which is representative of the larger dataset. The reports have been filtered to focus on the health sector and are `.txt` format.

**Reports' metadata:**

Additionnaly to the USAID evaluation reports, a previous university research conducted by Carlitz and Bloem (2023) at the University of Amsterdam, on which we will base our work, consituted metadata for each of the reports. It adds the political affiliations of the main governing bodies at the time of each report (Senate, President, and House of Congress) and the dataset was augmented with information on the primary subject of each report, the outcomes of the results sections, and relevant keywords and tags from the USAID Thesaurus


**Thesaurus keywords and politicization score:**

The keywords necessary to determine whether a report is politicized or not are selected from the USAID Thesaurus database from 2021 and are filtered to only contain keywords health-related to align with our main dataset. Each keyword was manually labeled to establish a politicization score ranging from 1 to 3 (integers)- where 3 indicates high politicization, 2 denotes interventions or targets that are broadly defined and could be interpreted variably, and 1 signifies minimal political relevance. This hand-labeling process, inherently subjective, was undertaken by two research assistants to ensure a more balanced assessment. A keyword is considered politicized if there is an ideological disagreement along the constituent line about the means to achieve a certain goal. For this reason, these labels serve as a measure to evaluate the degree of politicization across the dataset.

```python
# Function to select n first sentences from a text
def select_n_sentences(text:str, n:int=2):
  '''
    Returns the n first sentences of a text.

    Parameters:
        text: The string which is to be cropped.

    Returns:
        first_two_sentences:The string containing 2 sentences.
  '''
  sentences = re.split(r'(?<=[.!?]) +', text)
  first_n_sentences = ' '.join(sentences[:n])
  return first_n_sentences
```

```python
# Function to detect the language of a text
def detect_language_with_langdetect(row):
    '''
    Returns a tuple specifying the most probable language and the percentage of
    likelihood for the 'Text' column in a DataFrame row.

    Parameters:
        row: A Series object representing a DataFrame row.

    Returns:
        a tuple with the most probable language and probability
    '''
    try:
        langs = detect_langs(row['Text'])
        for item in langs:
            return item.lang, item.prob
    except:
        return "err", 0.0
```

```python
# we apply the function by row and assign the result to new columns
df[['lang', 'proba']] = df.apply(detect_language_with_langdetect, axis=1, \
                                 result_type='expand')

#Inspecting the shape
print(df.shape)
```

First we need a few steps before vectorizing the texts to be able to use `CountVectorizer` from Sklearn.

```python
# Function to perform basic cleaning steps to text data
def clean_text(text:str):
    '''
    Returns a string where we lowercased the text, and removed special
    characters and stopwords.

    Parameters:
        text: A string to clean.

    Returns:
        A lowecased text with special characters and stopwords removed.
    '''
    return " ".join(
        word for word in re.sub('[^a-zA-Z0-9-]+', ' ', text.lower()).split()
        if word not in stopwords.words('english')
    )
```

```python
# Applying the clean function to the text and keywords
# to standardize the strings
df['Text_clean'] = df['Text'].apply(clean_text)
df_keywords['Keyword_clean'] = df_keywords['Keyword'].apply(clean_text)
```

```python
# Function to create a corpus of words from text data
# we need to create a function to use .apply() per row
def split_text(text:str):
    '''
    Returns a list from all the words present in a certain string.

    Parameters:
        text: A text to tokenize.

    Returns:
        A list of words present in a text to create a corpus.
    '''
    return text.split()
```

We see that we have keywords with up to 6 ngrams. A method will be

```python
# Count number of ngrams we have in the keywords list
def count_all_keywords_ngrams(df:pd.DataFrame, column_name:str='Keyword_clean'):
    '''
    Returns a serie with the frequency of each ngram string present.
    We split each string of the column by space and count the number of words.

    Parameters:
        df: A df which contains a string column to process.
        column_name: A column's name to process

    Returns:
        A serie mapping each ngram to its frequency.
    '''
    return df[column_name].str.split().str.len().value_counts().sort_index()

# Count keywords with exactly 2 and 3 words
print(count_all_keywords_ngrams(df_keywords))
```

For a large number of keywords and texts, it's more efficient to use a vectorized approach that minimizes the number of scans through the texts. One common way to do this in Python is by using the `CountVectorizer` class from the scikit-learn library. This method will convert the text data into a matrix of token counts, which can then be used to count the occurrences of each keyword.

```python
# We use CountVectorizer with ngrams between 1 and 3 terms
vectorizer = CountVectorizer(ngram_range=(1, 6), \
                             vocabulary=list(set(df_keywords.Keyword_clean.to_list())))

# We fit the vectorizer to the text data and transform the texts into a count matrix.
X = vectorizer.fit_transform(df['Text_clean'])

# We convert the result to a DataFrame with named columns (for readability)
df_ngram_counts = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# We sum each column to get the total counts of each n-gram keyword
ngram_counts = df_ngram_counts.sum()

# ngram_counts is now a series with the counts of each n-gram keyword
print(ngram_counts)
```

```python
# Investigating the number of keywords not used
print(f"Number of keywords not used:{(ngram_counts==0).sum()}")
```

```python
# We count the number of ngrams appearance in the text
df_ngram = pd.DataFrame(ngram_counts, columns=['Occurrences'])

# Create a new column for the number of words in each keyword
df_ngram['Word_Count'] = df_ngram.index.str.split().str.len()

# Group by the number of words and sum the occurrences for each group
print(df_ngram.groupby('Word_Count')['Occurrences'].sum())
```

The following section focuses on comparing the keywords frequencies when we distinguish between reports evaluating interventions happenning under Republican or Democrat governments.

```python
def get_top_n_keywords(dataframe:pd.DataFrame, n:int=50):
    '''
    Returns a Dataframe with the top n keywords present in the input dataframe
    as well as their count.

    Parameters:
        dataframe: A Dataframe containing a column 'Text_clean' to analyze.
        n: An integer, the top n number of keywords we want to identify.

    Returns:
        A df with the top n keywords and their count.
    '''
    vectorizer = CountVectorizer(ngram_range=(1, 3), vocabulary=list(set(df_keywords.Keyword_clean.to_list())))

    X = vectorizer.fit_transform(dataframe['Text_clean'])
    df_ngram_counts = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

    ngram_counts = df_ngram_counts.sum()
    df = ngram_counts.sort_values(ascending=False).head(n).reset_index()

    df.columns = ['Keyword', 'Count']

    return df
```

```python
# Function to fill missing values of target columns
def fill_missing_values(df:pd.DataFrame, year_column:int, target_columns:list):
    '''
    Fills the missing values from the columns contained in target_colums based
    on the year_column value. The idea is that if we do not have these values,
    we'll take the information of another report which has the same year_column.
    This make sense since the information we are interested in are about
    governments and thus are the same for similar year (e.g. the president,
    senate and house of congress are the same in 2007, we do not take election
    years into account).
    Returns the filled df.

    Parameters:
        df: The dataframe of interest containing year_column and target_columns.
        year_column: The name of the column from which we fill the values of the
        target_columns.
        target_columns: A list containing the names of the columns to
        fill as str.

    Returns:
        The same df but with filled values for the target_columns.
    '''
    for index, row in df.iterrows():
        # we identify missing values
        missing_cols = [col for col in target_columns if pd.isnull(row[col])]

        if missing_cols:
            fill_row = df[(df[year_column] == row[year_column]) & df[missing_cols].notnull().all(axis=1)]
            if not fill_row.empty:
                for col in missing_cols:
                    df.at[index, col] = fill_row[col].values[0]
```

```python
# We apply the function to our dataset two times
# the first time for the columns related to 'Year Start'
# the second time for the columns related to 'Year End'
start_columns = ['President_Start', 'Senate_Start', 'House_Start', \
                 'PSH_Start', 'DivGov_Start']
end_columns = ['President_End', 'Senate_End', 'House_End', \
               'PSH_End', 'DivGov_End']

fill_missing_values(df, 'Year Start', start_columns)
fill_missing_values(df, 'Year End', end_columns)
```

### Distinguishing between governments

Until now we determined the government under which an intervention was conducted only on the party of the president at the start of an intervention. The reason was that it made sense that USAID interventions would be aligned with political agenda of the president, and it was a simplification that saved us some complications. Nevertheless, it might be too simple. Therefore we want to determine a better way to split the reports.

Evaluation reports asses interventions which usually spans over a few years. For this reason, the party an intervention was initally planned and conducted under can change. In order to determine whether a report is labeled as Democrat or Republican, we need to come up with a method to determine under which party the intervention was conducted the majority of the time.




#### Algo 1: _Institutional control_

Algorithm 1:

For reports which have values about the democratic fraction of the three institutions (`DemPres_FractionProject`, `DemSen_FractionProject`, `DemHouse_ProjectFraction`):
 we decide to determine that a project was conducted under Democrat government if 2 of these fractions are larger than 50%.

If we do not have value for the democratic fractions: we count the number of institutions belonging to each party at the end and at the beginning of the interventions. The one with the largest number is attributed as the `Institutional_Split` value.


```python
# Function to classify reports under a certain government
# rule based, could be changed easily
def split_institutional_control(row, target_columns:list = \
 ['President_Start', 'Senate_Start', 'House_Start',\
  'President_End', 'Senate_End', 'House_End']):
    '''
    Fills a row with either the value 'D' or 'R' to determine under which
    government a project was conducted the majority of the time. It takes
    two cases into account: whether we have access to information regarding
    Democrat fractions of the project or if we do not, in which case we compare
    the number of institutions belonging to each party at the begging and at the
    end of an intervention.

    Parameters:
        row: A row corresponding to a report.
        target_columns: A list of column names containing the party of
        the three institutions at the start and end of an intervention.

    Returns:
        A string value (either 'D' for Democrat or 'R' for Republican).
    '''
    # if we missed information about the fractions of the project
    if pd.isnull(row['DemPres_FractionProject']) or \
    pd.isnull(row['DemSen_FractionProject']) or \
     pd.isnull(row['DemHouse_ProjectFraction']):
        counts = {
            'D': sum(row[col] == 'D' for col in target_columns),
            'R': sum(row[col] == 'R' for col in target_columns)
        }
        return 'D' if counts['D'] > counts['R'] else 'R'
    # if fractions are available
    else:
        conditions_met = sum([
            row['DemPres_FractionProject'] > 0.5,
            row['DemSen_FractionProject'] > 0.5,
            row['DemHouse_ProjectFraction'] > 0.5
        ])
        return 'D' if conditions_met >= 2 else 'R'
```

```python
# Apply to each row of our dataframe
df['Institutional_Split'] = df.apply(split_institutional_control, axis=1)
```

We notice that there are twice the number of Republican reports compared to the ones of Democrat. This imbalance shouldn't impact the contextual embeddings methodology since we are not training a model with imbalanced data. Still, it could impact our results since the Republican reports are over-represented.

This is an issue for the future work. A potential solution would be to determine differently to what government a report belongs.

```python
# Inspecting the number of each value
df['Institutional_Split'].value_counts()
```

#### Algo 2: _Presidential control_

Algorithm 2:

In this case we want to split the report based only ont the president that was in place for the majority of the intervention.

For reports which have values about the democratic fraction of the president (`DemPres_FractionProject`):
 we determine that a project was conducted under Democrat government if the fraction is larger than 50%.

If we do not have value for the democratic fraction of president: we count the number of institutions belonging to each party at the end and at the beginning of the interventions (we can not simply base it on `President_Start` and `President_End` because these are only 2 values, which can create a tie case if they are from different parties). The one with the largest number is attributed as the `Presidential_Split` value.


We first need to assess if there are missing values in the values of interest.
No values are missing for the `President_Start` and `President_End`.

```python
print(df['President_Start'].isnull().sum())
print(df['President_End'].isnull().sum())
```

```python
def split_presidential_control(row, target_columns:list = \
 ['President_Start', 'Senate_Start', 'House_Start',\
  'President_End', 'Senate_End', 'House_End']):
    '''
    Fills a row with either the value 'D' or 'R' to determine under which
    government a project was conducted the majority of the time. It takes
    two cases into account: whether we have access to information regarding
    Democrat fraction of the project or if we do not, in which case we compare
    the number institutions for each party at the begging and at the end of
    an intervention.

    Parameters:
        row: A row corresponding to a report.
        target_columns: A list of column names containing the party of
        the three institutions at the start and end of an intervention.

    Returns:
        A string value (either 'D' for Democrat or 'R' for Republican).
    '''
    # if we missed information about the fractions of the project
    if pd.isnull(row['DemPres_FractionProject']):
        counts = {
            'D': sum(row[col] == 'D' for col in target_columns),
            'R': sum(row[col] == 'R' for col in target_columns)
        }
        return 'D' if counts['D'] > counts['R'] else 'R'
    # if fractions are available
    else:
        return 'D' if row['DemPres_FractionProject'] > 0.5 else 'R'
```

```python
df['Presidential_Split'] = df.apply(split_presidential_control, axis=1)
```

We notice that there is a gain a large imbalance between the reports classified as Republican and Democrat. This imbalance shouldn't impact the contextual embeddings methodology since we are not training a model with imbalanced data. Still, it could impact our results since the Republican reports are over-represented.
